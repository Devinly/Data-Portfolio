## **A1: Questions for Analysis** 
This study explores the relationship between being overweight and the likelihood of hospital readmission by using the research question, “Are patients who are overweight more prone to higher rates of readmission compared to the rest of the population?” The two variables were chosen because being overweight often coincides with other health conditions, creating a complex health profile. Given this, it is important to examine if there is a correlation between weight and readmission. 

This question is demonstrated with the null hypothesis as follows:

H0:readmissionOverweight=readmissionPopulation  
H1:readmissionOverweight≠readmissionPopulation

The alpha value of 0.05 (95% certainty) will be used.

## **A3: Data Identification** 
To answer the research question regarding being overweight and hospital readmission, the relevant data variables are as follows:

- Dependent Variable:
ReAdmis: Indicates whether the patient was readmitted within a month of release (Yes/No). 

- Independent Variable:
Overweight: Indicates whether the patient is overweight based on age, gender, and height (Yes/No).

## **B1: Code** 

```python
# Import libraries
import pandas as pd
from scipy.stats import chi2_contingency

# File path to CSV file
file_path = 'medical_clean.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)
```

```python
# Create a crosstab for overweight and readmission [In text citation: (Pandas crosstab(), n.d.)]
table = pd.crosstab(df['Overweight'], df['ReAdmis'])
print(table)
```

```python
# Perform chi-square test [In text citation: (Scipy.stats.chi2_contingency, n.d)]
chi_value, p_value, _, _ = chi2_contingency(table)

# Print chi-square and p-values
print(f"Chi-square value: {chi_value}")
print(f"P-value: {p_value}")

# Interpret results based on p-value
if p_value < 0.05:
    print("Reject the null hypothesis")
else:
    print("Accept the null hypothesis")
```

## **B2: Output**
A chi-square test for independence examined the association between being overweight and hospital readmission. The contingency table showed the count of patients categorized by their overweight status (No or Yes) and hospital readmission status (No or Yes). The test resulted in a chi-square value of approximately 0.70 and a p-value of 0.40. In hypothesis testing, “a p-value less than 0.05 is typically considered to be statistically significant, in which case the null hypothesis should be rejected” (Investopedia). Therefore, based on the findings, with an alpha level of 0.05 (95% confidence), the p-value of 0.40 is not below the threshold. There is not enough evidence to claim a significant difference, and thus, the null hypothesis is not rejected, which leads to the conclusion that there is no statistically significant association between being overweight and hospital readmission in the dataset. 

## **B3: Justification**
The chi-square test analyzes the association between two categorical variables. It assesses whether there is a significant association or independence between the variables. Therefore, the chi-square test suits this research because it is exploring the relationship between two binary, categorical variables: overweight and hospital readmission status. 

Chi-square tests are commonly used in healthcare research, where researchers often explore connections between different categories; they work well with nominal or ordinal categories and handle tables with moderately sized frequencies. Therefore, this test was chosen because it is a widely accepted statistical method in healthcare research and is also simple to use and understand. While there are other effect methods, this method is straightforward and aligns well with the study’s goal: to explore the connection between being overweight and the likelihood of hospital readmission.

```python
# Create scatter plot for TotalCharge and Income [In text citation: (Seaborn.scatterplot., n.d.)]
plt.title('TotalCharge vs Income (Bivariate, Continuous Variables)')
sns.scatterplot(data=df, x='TotalCharge', y='Income', color='blue')
plt.xlabel('TotalCharge')
plt.ylabel('Income (USD)')
```

```python
# Create stacked bar graph for HighBlood and Complication_risk

data = {
    'HighBlood': ['yes', 'no', 'yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no'],
    'Complication_risk': ['high', 'medium', 'low', 'high', 'medium', 'low', 'low', 'high', 'medium', 'low']
}
df = pd.DataFrame(data)
ct = pd.crosstab(df['HighBlood'], df['Complication_risk'])
ct = ct[['high', 'medium', 'low']]
ax = ct.plot(kind='bar', stacked=True, color=['red', 'skyblue', 'green'], alpha=0.7)
plt.xlabel('HighBlood')
plt.ylabel('Count')


plt.title('HighBlood vs Complication Risk (Bivariate, Categorical Variables)')
plt.legend(title='Complication Risk', loc='upper right', bbox_to_anchor=(1.25, 1))

plt.tight_layout() 
plt.show()
```

## **E1: Results of Analysis** 
As previously mentioned, the results of the chi-square test for independence show a chi-square value of 0.70 and a p-value of approximately 0.40. The null hypothesis, which assumed no significant association between being overweight and hospital readmission, was not rejected. The p-value is particularly higher than the alpha level of 0.05, meaning that there is not enough evidence to reject the null hypothesis. In practical terms, this implies that, based on the analyzed data, being overweight does not seem to have a notable impact on the likelihood of hospital readmission. The relationship between these two things is not different enough from what would be expected if they were independent. As a result, there is insufficient evidence to support the notion that being overweight significantly increases the chance of hospital readmission in the group studied. 

## **E2. Limitations of Analysis**
The analysis, while informative, has some limitations that should be kept in mind. First and foremost, the chi-square test has limitations in itself. It is a method that helps determine if there is a relationship or correlation between two categorical variables; however, it cannot prove causation. Therefore, other factors like patient behaviors or lifestyle changes not within the data might influence hospital readmission. Furthermore, the dataset might miss important details that could be helpful in understanding more about hospital readmission, like specific treatments and post-discharge plans. Due to this, it is important to be careful when interpreting the results. 

Another limitation of the analysis is regarding the dataset size and composition. The dataset used had 10,000 patient records. This sample size is not particularly large and does not represent all kinds of hospitals; a bigger and more diverse dataset could provide more insights into this relationship. Moreover, changes over time are not a factor in this analysis. In healthcare, practices, patient demographics, etc., change over time, which may influence the results. In conclusion, while the chi-square test provides helpful insights, the limitations highlight the importance of carefully interpreting results. It suggests consideration of these drawbacks and encourages further, more detailed studies to better understand the connection between being overweight and hospital readmissions.
